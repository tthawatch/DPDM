```
from sklearn.model_selection import train_test_split
#แบ่งข้อมูลเป็น train กับ test แล้วเก็บ test เอาไว้
```

```
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2) # test เป็น 20 % train 80 %
```

```
print(f'{X_train.shape} {X_test.shape}')
```

```
#ทำ k-fold cross validation เพื่อเลือก model ที่ดีที่สุด
from sklearn.model_selection import cross_val_score
```

```
# 1. Import
from sklearn.neighbors import KNeighborsClassifier
# 2. Define model
KNN_k3_wU = KNeighborsClassifier(n_neighbors=3)
# 3. Cros validation
cross_val_score(KNN_k3_wU, X_train, Y_train, cv=5).mean() #นำเทสมาเฉลี่ย
```

```
# 1. Define model
KNN_k3_wD =KNeighborsClassifier(n_neighbors=3,weights='distance')
# 2. Cros validation
cross_val_score(KNN_k3_wD, X_train, Y_train, cv=5).mean()
```

```
# 1. Define model
KNN_k3_wE =KNeighborsClassifier(n_neighbors=5)
# 2. Cros validation
cross_val_score(KNN_k3_wE, X_train, Y_train, cv=5).mean()
```

```
# 1. Define model
KNN_k3_wF =KNeighborsClassifier(n_neighbors=5,weights='distance')
# 2. Cros validation
cross_val_score(KNN_k3_wF, X_train, Y_train, cv=5).mean()
```

```
# 1. Define model
KNN_k3_wG =KNeighborsClassifier(n_neighbors=10)
# 2. Cros validation
cross_val_score(KNN_k3_wG, X_train, Y_train, cv=5).mean()
```

```
# 1. Define model
KNN_k3_wF =KNeighborsClassifier(n_neighbors=10,weights='distance')
# 2. Cros validation
cross_val_score(KNN_k3_wF, X_train, Y_train, cv=5).mean()
```

Train model ที่ดีที่สุดด้วย training set แล้ววัดผลด้วย test set

```
#Define
KNN_k3_wG =KNeighborsClassifier(n_neighbors=3,weights='distance')
#train
KNN_k3_wG.fit(X_train,Y_train)
#Test_Evaluation
KNN_k3_wG.score(X_test,Y_test)
```
