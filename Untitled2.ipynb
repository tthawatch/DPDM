```
#ทำ k-fold cross validation เพื่อเลือก model ที่ดีที่สุด
from sklearn.model_selection import cross_val_score
```

```
# 1. Import
from sklearn.neighbors import KNeighborsClassifier
# 2. Define model
KNN_k3_wU = KNeighborsClassifier(n_neighbors=3)
# 3. Cros validation
cross_val_score(KNN_k3_wU, X_train, Y_train, cv=5).mean() #นำเทสมาเฉลี่ย
```

```
# 1. Define model
KNN_k3_wD =KNeighborsClassifier(n_neighbors=3,weights='distance')
# 2. Cros validation
cross_val_score(KNN_k3_wD, X_train, Y_train, cv=5).mean()
```

```
# 1. Define model
KNN_k3_wE =KNeighborsClassifier(n_neighbors=5)
# 2. Cros validation
cross_val_score(KNN_k3_wE, X_train, Y_train, cv=5).mean()
```

```
# 1. Define model
KNN_k3_wF =KNeighborsClassifier(n_neighbors=5,weights='distance')
# 2. Cros validation
cross_val_score(KNN_k3_wF, X_train, Y_train, cv=5).mean()
```

```
# 1. Define model
KNN_k3_wG =KNeighborsClassifier(n_neighbors=10)
# 2. Cros validation
cross_val_score(KNN_k3_wG, X_train, Y_train, cv=5).mean()
```

```
# 1. Define model
KNN_k3_wF =KNeighborsClassifier(n_neighbors=10,weights='distance')
# 2. Cros validation
cross_val_score(KNN_k3_wF, X_train, Y_train, cv=5).mean()
```
