##Generate Data

```
from sklearn.datasets import make_blobs
```

```
X, Y  = make_blobs(n_samples=300, centers=3, n_features=2, random_state=6) #สร้างข้อมูล 300 จุด 3 กลุ่มแต่ละข้อมูล 2 ลักษณะ
```

```
from matplotlib import pyplot as plt
plt.scatter(X[:,0],X[:,1])
```

##Import

```
from sklearn.cluster import KMeans
```

```
KMclust = KMeans(n_clusters=3)
```

```
KMclust.fit(X)
```

```
label = KMclust.predict(X)
```

```
plt.scatter(X[:, 0], X[:, 1], c=label)
```

```
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```
import numpy as np
```

```
np.unique(RGB_img)
```

```
#Define
myQuantize = KMeans(n_clusters=6)
```

```
img_row = np.reshape(RGB_img,(750*500,3))
img_row
```

```
#train
myQuantize.fit(img_row)
```

```
#predict
res = myQuantize.predict(img_row)
```

```
res
```

```
Clus_cen = myQuantize.cluster_centers_
Clus_cen
```

```
Clus_cen.shape
```

```
new_img = np.zeros(RGB_img.shape)
for i in range(RGB_img.shape[0]):
    for j in range(RGB_img.shape[1]):
        new_img[i,j,:] = Clus_cen[myQuantize.predict([RGB_img[i,j,:]]),:]
```

```
new_img_ = new_img/255
```

```
plt.imshow(new_img_)
```
